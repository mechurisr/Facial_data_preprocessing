```python
import pandas as pd
import numpy as np
from os import listdir
from os.path import isfile, join


files = [f for f in listdir('googlCrawlingEmotionDataset2/angry_color') if isfile(join('googlCrawlingEmotionDataset2/angry_color', f))]
files
```

```python
df = pd.DataFrame(files)
df
```

```python
df["Facial_Emotion"] = "angry"
df
```

```python
df.rename(columns = {0 : "Filepath"}, inplace=True)
df
```

```python
df.index.name="Num"
df
```

```python
df=df.set_index("Filepath")
df
```

```python
df
```

```python
sadfiles = [f for f in listdir('sad_color') if isfile(join('sad_color', f))]
sadDF = pd.DataFrame(sadfiles)
sadDF.rename(columns = {0 : "Filepath"}, inplace=True)
sadDF["Facial_Expression"] = "sad"
sadDF=sadDF.set_index("Filepath")
sadDF
```

```python
sadDF
```

```python
hapfiles = [f for f in listdir('happy_color') if isfile(join('happy_color', f))]
hapDF = pd.DataFrame(hapfiles)
hapDF.rename(columns = {0 : "Filepath"}, inplace=True)
hapDF["Facial_Expression"] = "happy"
hapDF=hapDF.set_index("Filepath")
hapDF
```

```python
neufiles = [f for f in listdir('neutral_color') if isfile(join('neutral_color', f))]
neuDF = pd.DataFrame(neufiles)
neuDF.rename(columns = {0 : "Filepath"}, inplace=True)
neuDF["Facial_Expression"] = "neutral"
neuDF=neuDF.set_index("Filepath")
neuDF
```

```python
faceData = pd.concat([df,neuDF,sadDF,hapDF])
faceData
```
